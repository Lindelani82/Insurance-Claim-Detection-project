## Data Cleaning

```python
new_data.isnull().sum()
```

```python
new_data.drop("_c39", axis=1, inplace=True)
```

```python
new_data
```

```python
new_data.duplicated().sum()
```

```python

# Select numerical columns
numerical_columns = new_data.select_dtypes(include=['float64', 'int64']).columns

# Determine grid size
n_cols = 3  # Number of plots per row
n_rows = math.ceil(len(numerical_columns) / n_cols)

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))  # Adjust size
axes = axes.flatten()  # Flatten axes for easy iteration

# Plot each column
for i, column in enumerate(numerical_columns):
    sns.boxplot(data=new_data, x=column, ax=axes[i])
    axes[i].set_title(f'Boxplot for {column}')
    axes[i].set_xlabel('')  # Remove x-label to reduce clutter

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
